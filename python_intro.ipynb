#### Matrix addition

A function addMatrix(matA, matB) which returns a new matrix after addition


````python
matA = [
    [ 0, 1, 2],
    [ 2, 3, 5],
    [ 4, 5, 2],
    [ 5, 6, 4]
]

matB = [
    [ 1, 2, 4],
    [ 4, 5, 3],
    [ 4, 6, 2],
    [ 2, 4, 1]
]
matC = []
for i in range(len(matA)):
    row = []
    for j in range(len(matA[0])):
      row.append(matA[i][j] + matB[i][j])
    matC.append(row)

def displayMatrix(mat):
   for row in mat:
    for col in row:
      print(col, end=' ')
    print()
displayMatrix(matC)



````

Let's build a `Caesar Cipher`

Hint:
```python
ord('a') # gives ascii value
chr(97) # gives character, here, 'a'
```


````python
def cipher(word , shift_count):
  result = ""
  for char in word.lower():
     # skipping those iterations which points to any other characters than alphabets
    if ord(char) < 97 or ord(char) > 122:
      result += char
      continue # skips the rest of current iteration
    shifted_ascii = ord(char) + shift_count
    if shifted_ascii >= 122:
      shifted_ascii = 97 + (shifted_ascii - 123)
      result += chr(shifted_ascii)
    else:
      result += chr(ord(char)+shift_count)
  return result

note = "hi i'm vinith"
ciphered_message = cipher(note,2)
print(ciphered_message)

````

````python
def decipher(word , shift_count):
  result = ""
  for char in word.lower():
    shifted_ascii = ord(char)-shift_count
     # skipping those iterations which points to any other characters than alphabets
    if ord(char) < 97 or ord(char) > 122:
      result += char
      continue
    shifted_ascii = ord(char) - shift_count
    if shifted_ascii < 97:
      shifted_ascii = 123 - (97-shifted_ascii)
      result += chr(shifted_ascii)
    else:
      result += chr(ord(char) - shift_count)
  return result

decode = "jk k'o xkpkvj"
deciphered_message = decipher(decode,2)
print(deciphered_message)

````

Take the person's age, check if the person is adult.

If adult, then show `Hi {name} you are {age} years old and you can drive`

else if not adult, then show `Hi dear {name}, you have to wait <no_of_years> years to start driving`

For eg :

If the name is Ragul and  age is 17,

We will show
`Hi dear Ragul, you have to wait 1 year to start driving`


````python
AGE_LIMIT = 18
name = input("Enter your name : ")
try:
  age  = int(input("Enter your age  : "))
except:
  print("Wrong input")

if age >= AGE_LIMIT:
  print(f"Hi {name} you are {age} years old and you can drive")
else:
  diff = AGE_LIMIT - age
  year_word = "years" if diff > 1 else "year"
  print(f"Hi dear {name}, you have to wait {diff} {year_word} to start driving")

````

````python
# functions to handle the question bank
QUESTION_BANK_FILE_NAME = 'questions.json'

def questions_to_file(question_bank, file_name):
  """
  This function writes the given question bank to the provided file_name
  in JSON format.
  """
  import json
  with open(file_name, 'w') as questions_file:
    json.dump(question_bank, questions_file, indent=4)

questions_to_file(question_bank, QUESTION_BANK_FILE_NAME)


q21 = {
    'id': 21,
    'q': "Which method is used to add an element at the end of a list in Python?",
    'choices': ['append()', 'add()', 'insert()', 'push()'],
    'correct_index': 0,
    'hint': "It appends the item to the list."
}

def read_questions_from_file(file_name):
  """
  Reads question bank from the provided json file to the global question_bank
  variable.
  """
  import json
  with open(file_name, 'r') as questions_file:
    global question_bank
    question_bank = json.load(questions_file)
    # this converts and returns the list of dictionary structure into memory
  return question_bank

# I want to add this to the file
def add_question_to_file(question, file_name):
  """
  Appends a new question into the file as well as updates the
  global question_bank variable in the process.

  Question must be of format :
  {
    'id' : 1,
    'q' : "Some question?",
    'choices' : ['choice1', 'choice2', 'choice3', 'choice4' ],
    'correct_index' : 0,
    'hint' : "some hint"
  },
  """

  # 1. Read existing file
  read_questions_from_file(file_name)
  # overwrites the global question bank from file

  # 2. Append the new dictionary
  global question_bank
  question_bank.append(question)
  # 3. Write back new list to json file
  questions_to_file(question_bank, file_name)

add_question_to_file(q21, QUESTION_BANK_FILE_NAME)
````

````python
def main_menu():
  prompt = """
  Please select your choice:
  1. View question bank
  2. Add question
  0. Exit
  """

  choice = input(prompt)
  try:
    choice = int(choice)
  except ValueError:
    print("Please check your input")
    main_menu()



  match choice:
    case 1:
      print("View question bank")
    case 2:
      print("Add question")
    case 0:
      print("Exit")
    case _:
      print("Please check your input")
main_menu()
````
